```python
import pandas as pd
import numpy
import requests
import json
import re
```

```python
df = pd.read_csv("all_data.csv")
display(df)
```

```python
channelId = []
for video in df['videoid']:
    r = requests .get('https://www.googleapis.com/youtube/v3/videos?part=snippet&id='
                      + video + 
                      '&key=AIzaSyBGchg4qx7BerIXg8OMi9g5s3cW76e0eqs')
    json_dict = json .loads(r.text)                 
    channelId.append(json_dict['items'][0]['snippet']['channelId'])
```

```python
subscriberCount = []
for channel in channelId:
    r = requests .get('https://www.googleapis.com/youtube/v3/channels?part=statistics&id='
                      + channel + 
                      '&key=AIzaSyBGchg4qx7BerIXg8OMi9g5s3cW76e0eqs')
    json_dict = json .loads(r.text)
    if 'items' in json_dict:
        subscriberCount.append(int(json_dict['items'][0]['statistics']['subscriberCount']))
    else:
        subscriberCount.append(numpy.nan)
```

```python
df = pd.concat([df,pd.Series(subscriberCount)],axis=1)
df.columns.values[18] = 'subscriberCount'
```

```python
import pandas as pd
import requests
import json
import re
```

```python
### read in the csv
df = pd.read_csv("df_all_raw.csv")
```

```python
display(df)
```

```python
viewCount = []
for video in df['videoid']:
    r = requests .get('https://www.googleapis.com/youtube/v3/videos?part=statistics&id='
                      + video + 
                      '&key=AIzaSyBGchg4qx7BerIXg8OMi9g5s3cW76e0eqs')
    json_dict = json .loads(r.text)                 
    viewCount.append(int(json_dict['items'][0]['statistics']['viewCount']))
```

```python
for i in range(len(viewCount)):
    viewCount[i] = int(viewCount[i])
```

```python
df.viewcount = viewCount
```
